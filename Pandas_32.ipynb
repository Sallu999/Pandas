# #32: Plotting dari suatu Pandas Data Frame

## Import Modules

```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

## Persiapan Data Frame

```python
n_rows = 40
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), 
                  columns=cols)
df.head()
```

## Line Plot

```python
df.head()
```

```python
df.plot(kind='line')
```

```python
df[['A', 'B']].plot(kind='line')
```

## Bar Plot

```python
df.head()
```

```python
df.plot(kind='bar')
```

```python
df[['A', 'B']].plot(kind='bar')
```

```python
df[['A', 'B']].head().plot(kind='bar')
```

```python
df[['A', 'B']].head().plot(kind='barh')
```

## Area Plot

```python
df.head()
```

```python
df.plot(kind='area')
```

```python
df[['A', 'B']].head().plot(kind='area')
```

## Box Plot

```python
df.head()
```

```python
df.plot(kind='box')
```

## Histogram

```python
df.head()
```

```python
df.plot(kind='hist')
```

```python
df[['A', 'B']].plot(kind='hist')
```

## Kernel Density Estimation (KDE)
